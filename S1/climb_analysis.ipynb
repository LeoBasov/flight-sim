# Influence of $c_D \cdot S$ on max altitude

```python
W0 = 10000
Wf = 6500
nu_b = 0.9

cdS = np.linspace(1, 2)

asc.set_mass(W0, Wf, nu_b)
h_max = []

for c in cdS:
    asc.set_cd_S(c, 1.0)
    asc.run()
    h_max.append(asc.h[-1])
    
plt.plot(cdS, h_max)
plt.show()
```
